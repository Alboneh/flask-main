```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
for column in data.columns:
    print(column)
    print(data[column].unique().shape, "\n")
```

```python
data = data.loc[:, ["Date", "itemDescription"]]
data.head()
```

```python
data["Date"] = pd.to_datetime(data["Date"], format="%d-%m-%Y")
```

```python
data["count"] = 1
```

```python
data.head()
```

```python
grouped_data = data.groupby(["Date", "itemDescription"]).sum()
grouped_data
```

```python
grouped_data.reset_index(inplace = True)
```

```python
grouped_data.head(10)
```

```python
sparse_data = pd.pivot_table(grouped_data, values = "count", columns = "itemDescription", index = "Date", aggfunc = np.sum)
sparse_data.head()
```

```python
sparse_data = sparse_data.applymap(lambda x: 0 if np.isnan(x) else x)
sparse_data
```

```python
for i in sparse_data.columns:
    print(i)
```

```python
columns_to_not_be_removed = [
    "beef",
    "berries",
    "brown bread",
    "chicken",
    "citrus fruit",
    "condensed milk",
    "fish",
    "fruit/vegetable juice",
    "grapes",
    "ham",
    "hamburger meat",
    "herbs",
    "honey",
    "meat",
    "onions",
    "rice",
    "spices",
    "turkey",
    "waffles",
    "white bread",
    "yogurt"
]
```

```python
for column in sparse_data.columns:
    if column not in columns_to_not_be_removed:
        sparse_data.drop(column, axis = 1, inplace = True)
```

```python
for column in columns_to_not_be_removed:
    if column not in sparse_data.columns:
        sparse_data.drop(column, axis = 1, inplace = True)
```

```python
print(sparse_data.shape)
sparse_data.head()
```

```python
sparse_data.to_csv("sparse_data.csv")
```
